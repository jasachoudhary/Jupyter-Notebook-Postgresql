```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to the PostgreSQL database
db = create_engine('postgresql://myuser:mypass@postgres/mydb')

# Load data from PostgreSQL into a pandas DataFrame
df = pd.read_sql_query('SELECT * FROM trade.trades', con=db)

# Calculate number of day trades per account
day_trades = df.groupby('account_id').size().reset_index(name='day_trade_count')

# Save the result to a CSV file
day_trades.to_csv('output_day_trades.csv', index=False)

db.dispose()
```
